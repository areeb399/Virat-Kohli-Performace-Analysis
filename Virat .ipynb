```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Virat.csv")
df
```

```python
df.dropna(inplace = True)
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Changing the data type

df.runs = df.runs.astype(int)
df['date'] = pd.to_datetime(df['date'])

# Add a new column for year
df['year'] = df['date'].dt.year

# Add a new column for month
df['month'] = df['date'].dt.month
```

```python
df.head()
```

```python
#Total Runs Scored 

total_runs =df['runs'].sum()

print("Total Runs Scored : ", total_runs)
```

```python
# Calculate the average runs scored
average_runs = df['runs'].mean()

print("Average Runs Scored:", average_runs)
```

```python
# Find the maximum runs scored
max_runs = df['runs'].max()

print("Maximum Runs Scored:", max_runs)
```

```python
#Grouping the data by the 'ground' column and then calculating the sum of runs for each group

Ground = df.groupby('ground')['runs'].sum().reset_index()

print(Ground)
```

```python
#Creating a bar plot using seaborn (sns) for grounds based on the runs scored

plt.figure(figsize = (20,14))

sns.barplot(x = Ground['ground'], y = Ground['runs'])

plt.xticks(rotation = 90)

plt.xlabel('Ground', fontweight = 'bold')
plt.ylabel('Runs', fontweight = 'bold')
plt.title('Runs Scored on Different Grounds', fontweight = 'bold')

plt.show()
```

```python
# Sort the DataFrame in descending order based on the 'runs' column
Ground_sorted = Ground.sort_values(by='runs', ascending=False)

# Get the top 5 rows
top_10_values = Ground_sorted.head(10)

# Print the result
print(top_10_values)
```

```python
#Creating a bar plot using seaborn (sns) for the top 10 grounds based on the runs scored

plt.figure(figsize=(12, 8))

fig = sns.barplot(x = 'ground' , y = 'runs' , data = top_10_values)

for bars in fig.containers:
    fig.bar_label(bars)

plt.xticks(rotation = 90)

plt.xlabel('Ground', fontweight = 'bold')
plt.ylabel('Runs', fontweight = 'bold')
plt.title('Top 10 Runs by Ground', fontweight = 'bold')

plt.show()
```

```python
Conclusion:
    The maximum runs scored in Mirpur Ground is 1258
```

```python
#Grouping the data by the 'opponent' column and calculates the sum of runs scored against each opponent

Opponent = df.groupby('opponent')['runs'].sum().reset_index()

print(Opponent)
```

```python
#Creating a bar plot using seaborn (sns) for the opponents and their corresponding runs

plt.figure(figsize = (14,10))

fig = sns.barplot(x = Opponent['opponent'], y = Opponent['runs'])

for bars in fig.containers:
    fig.bar_label(bars)

plt.xticks(rotation = 90)

plt.xlabel('Opponents', fontweight = 'bold')
plt.ylabel('Runs', fontweight = 'bold')
plt.title('Runs against Opponent', fontweight = 'bold')

plt.show()
```

```python
# Sort the DataFrame in descending order based on the 'runs' column
opponent_sorted = Opponent.sort_values(by='runs', ascending=False)

# Get the top 5 rows
top_5_values = opponent_sorted.head(5)

# Print the result
print(top_5_values)
```

```python
#Creating a bar plot using seaborn (sns) for the top 5 opponents based on the runs scored

plt.figure(figsize = (14,10))

fig = sns.barplot(x = top_5_values['opponent'], y = top_5_values['runs'])

for bars in fig.containers:
    fig.bar_label(bars)


plt.xlabel('Opponents', fontweight = 'bold')
plt.ylabel('Runs', fontweight = 'bold')
plt.title('Runs against Opponent', fontweight = 'bold')

plt.show()
```

```python
Conclusion : 
    The sum of runs scored against Australia are 4483, followed by 3903 against England, and then 3653 runs as the third highest.
```

```python
# Calculate the maximum runs scored against each opponent
max_runs = df.groupby('opponent')['runs'].max().reset_index()

# Sort the DataFrame by maximum runs in descending order
max_runs = max_runs.sort_values(by='runs', ascending=False)

print(max_runs)
```

```python
#Creating a bar plot in seaborn (sns) that shows the maximum runs scored against each opponent team

plt.figure(figsize=(12, 6))

fig = sns.barplot(data=max_runs, x='opponent', y='runs', palette='viridis')

for bars in fig.containers:
    fig.bar_label(bars)


plt.title('Maximum Runs Scored Against Each Opponent', fontweight = 'bold')
plt.xlabel('Opponent Team' , fontweight = 'bold')
plt.ylabel('Maximum Runs' , fontweight = 'bold')


plt.xticks(rotation=45, ha = 'right')


plt.show()
```

```python
Conclusion:
    The maximum runs scored against South Africa are 254, followed by 243 against Sri Lanka, and then 235 against England as the third highest.
```

```python
#Grouping the data by the 'year' column and calculates the sum of runs for each year

yearwise = df.groupby('year')['runs'].sum()

print(yearwise)
```

```python
#Creating a bar plot using seaborn (sns) for the yearwise distribution of runs

plt.figure(figsize = (15,8))

fig = sns.barplot(x = yearwise.index, y = yearwise.values)

    
for bars in fig.containers:
    fig.bar_label(bars)

plt.xticks(rotation = 90)
plt.xlabel('Years', fontweight = 'bold')
plt.ylabel('Total Runs', fontweight = 'bold')
plt.title('Yearwise Distribution of Runs', fontweight = 'bold')

plt.show()    
    
```

```python
Conclusion:
    The highest runs scored in the year 2017 was 2818.
```

```python
#Creating a pie chart for the yearwise distribution of runs

plt.figure(figsize=(18, 10))


yearwise.plot(kind='pie', autopct='%0.1f%%', startangle=135)


plt.title("Yearwise Distribution of Runs", fontweight='bold')

plt.show()
```
